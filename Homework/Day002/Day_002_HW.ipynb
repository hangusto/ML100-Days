#### 找出 Total Income的最大值、最小值、中位數、平均值，同時記錄數值的位置，並自行驗證答案

```python
total_income = app_train.loc[:,'AMT_INCOME_TOTAL']

total_income_MAX = total_income.max()
total_income_MAX_LOC = total_income.idxmax()
total_income_MIN = total_income.min()
total_income_MIN_LOC = total_income.idxmin()
total_income_MED = total_income.median()
total_income_AVG = total_income.mean()

print('% Statistics of Total Income')
print('Max: %d [%d]' %(total_income_MAX, total_income_MAX_LOC))
print('Min: %d [%d]' %(total_income_MIN, total_income_MIN_LOC))
print('Med: %d' %(total_income_MED))
print('Avg: %f' %(total_income_AVG))
print('---')
print('% Data Check')
print(total_income.describe())
```
